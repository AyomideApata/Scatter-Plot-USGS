##  Working on Data
1. Created a clean database, remove all the formation with not a valid name: 'UNKNOWN' and 'NaN'
2. Removed the coloumns: Twn Num, Twn Dir, 	Rng Num, Rng Dir, Sec,	Quarter and Lib Num.

```python
## Saved the cores in a data frame name df, remove all the rows where the formation name is 'UNKNOWN'

df=cores[cores['Formation']!=('UNKNOWN')]
df
```

```python
## Remove the coloumns: Operator,Twn Num, Twn Dir, Rng Num,County Thin Sec, Analysis, Rng Dir, Sec, Quarter, Lib Num, Field, Type, Photos,  from df
## Changing dataframe to df_cores_clean

df_cores_clean=df.drop(columns=['Operator','Twn Dir', 'Rng Num', 'County','Rng Dir','Thin Sec', 'Analysis', 'Sec', 'Quarter', 'Lib Num', 'Field', 'Type', 'Photos'])
df_cores_clean
```

### Task 2
 
1. Use the df_cores_clean, and create a thickness coloumn in this dataframe, this can be computed by computing Max Depth - Min Depth
2. Createing a new dataframe (df_cores_clean_only_thick), with all the cores more than 200 ft core.
3. Finding the Top 10 Formations with the most cores and plot them with Seaborn
4. Finding the Thickness Histogram for the GREEN RIVER formation
5. Groupby 'State', 'Formation', and using the aggregate function to look at min', 'max', 'count'


```python
##this can be computed by computing Max Depth - Min Depth

df_cores_clean['Thickness']=df_cores_clean['Max Depth']-df_cores_clean['Min Depth']
df_cores_clean
```

```python
## df_cores_clean_only_deep=df_cores_clean[df_cores_clean['Min Depth']>5000.0]
# df_cores_clean_only_thick=df_cores_clean[(df_cores_clean['Thickness']>200.0)& (df_cores_clean['Min Depth']>5000.0)]

df_cores_clean_only_thick=df_cores_clean[df_cores_clean['Thickness']>200.0]
df_cores_clean_only_thick
```

### Import matplotlib and seaborn. 

```python
import seaborn as sns
import matplotlib.pyplot as plt

counts=df_cores_clean_only_thick['Formation'].value_counts()
## Create a chart from seaborn
sns.barplot(y=counts.index[0:10], x=counts.values[0:10])

plt.title('Top 10 Formations with the most cores')
plt.show()
```

```python
## This is all the Texas Cores
df_cores_clean_only_thick[df_cores_clean_only_thick['State']=='TX'].Thickness.hist()
```

```python
##Find the Thickness Histogram for the GREEN RIVER formation ( see the example above)

df_cores_clean_only_thick[df_cores_clean_only_thick['Formation']=='GREEN RIVER' ].Thickness.hist()
```

```python
## Groupby 'State', 'Formation', and using the aggregate function to look at min', 'max', 'count'

df_cores_clean_only_thick.groupby(['State', 'Formation']).agg(['min', 'max', 'count'])
```

### Task 3
Add Numpy, Seaborn and Matplotlib libraries. 
Creating graph (Scatter plot) using Matplotlib. RHOB, NPHI & DT
RHOB -Bulk Density 
NPHI -Neutron Pororsity

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
##Import the RandomWell.csv Well from DATA folder

well = pd.read_csv("/Users/Ayomide/Downloads/Python Github Excerise/DATA/RandomWell.csv")
well.head()
```

```python
##Plot RHOB curve, x should be RHOB and y should be DEPTH; limit RHOB values between 2 and 3, where as Depth between 4600, 4800 and DEPTH should be plotted in increasing downward

well.plot(kind = 'line', x = 'RHOB', y = 'DEPTH', figsize=(5, 10), xlim=(2, 3), ylim=(4800, 4600))
```

```python
## Createing a scatter plot for  x=RHOB and y= DT, and color it with GR with a color map of viridis_r 
## MatplotLib wasnt used

well.plot(kind = 'scatter', x = 'RHOB', y = 'DT', c='GR', colormap='viridis_r', figsize=(7, 6))
```

```python
# Createing a scatter plot for  x=NPHI and y=RHOB. 
fig, ax = plt.subplots(figsize=(3,2), dpi=200.0)

fig.suptitle('NPHI vs RHOB  crossplot')

im=ax.scatter(x=well['NPHI'], y=well['RHOB'],c=well['GR'], marker='o',cmap='viridis_r')
ax.set_xlabel("NPHI")
ax.set_ylabel("RHOB")

fig.colorbar(im, ax=ax)
im.set_clim(0.0, 100) 
plt.show()
fig.savefig("NPHI_vs_RHOB.png")
```

```python
# Createing a scatter plot for  x=RHOB and y= DT, and coloring it with GR with a color map of viridis_r 
## MatplotLib was used to see the example above

fig, ax = plt.subplots(figsize=(3, 2), dpi=200.0)

fig.suptitle('RHOB vs DT  crossplot')

im=ax.scatter(x=well['RHOB'], y=well['DT'],c=well['GR'], marker='o',cmap='viridis_r')
ax.set_xlabel("RHOB")
ax.set_ylabel("DT")

fig.colorbar(im, ax=ax)
im.set_clim(0.0, 100) ## Set color plot limits between 0 to 100 API
plt.show()
fig.savefig("RHOB_vs_DT.png")
```
